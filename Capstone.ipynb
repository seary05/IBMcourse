```python
import os
from google.colab import userdata

api_token = userdata.get("api_token")
os.environ["REPLICATE_API_TOKEN"] = api_token
```

```python
from langchain_community.llms import Replicate

parameters ={
    "top_k":5,
    "top_p":1.0,
    "max_tokens":4096,
    "min_tokens":0,
    # "max_length": 200,
    "random_speed": None,

    "repetition_penalty": 1.0,
    "temperature": 0.7,
    "stopping_criteria": "lenght (256 tokens)",
    "stopping_sequence": None
}

llm = Replicate(
    model = "ibm-granite/granite-3.3-8b-instruct",
    input = parameters,
)
```

```python
import pandas as pd

df = pd.read_csv("MobileDevice.zip")

df
```

```python
df_android = df[df["Operating System"] == "Android"]
df_android
```

```python
df_ios = df[df["Operating System"] == "iOS"]
df_ios
```

```python
from langchain_experimental.agents.agent_toolkits.pandas.base import (
    create_pandas_dataframe_agent
)

import pandas as pd

```

```python
agent = create_pandas_dataframe_agent(
    llm=llm,
    df=df,
    verbose=True,
    allow_dangerous_code=True,
    handle_parsing_errors=True,
    )

agent_android = create_pandas_dataframe_agent(
    llm=llm,
    df=df_android,
    verbose=True,
    allow_dangerous_code=True,
    handle_parsing_errors=True,
    )

agent_ios = create_pandas_dataframe_agent(
    llm=llm,
    df=df_ios,
    verbose=True,
    allow_dangerous_code=True,
    handle_parsing_errors=True,
    )
```

```python
agent.invoke({"input": "create pie plot between 'Android' and 'IOS' in 'Operating System'"})
```

Dari pie chart diatas dapat kita lihat bahwa mayoritas orang didalam data menggunakan android

```python
agent.invoke({"input": "Draw a chart bar from the count of first word in 'device model'"})
```
